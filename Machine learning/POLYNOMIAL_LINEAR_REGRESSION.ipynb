```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
df = pd.read_csv('Position_Salaries.csv')
df
```

```python
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
y
```

```python
X.reshape(1,-1)
y.reshape(1,-1)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
new_LR = LinearRegression()
new_LR.fit(X_poly,y)
```

```python
sn.set()
```

```python
plt.scatter(X,y,color='r',label='Scatter_values')
plt.plot(X,LR.predict(X),color='b',label='line')
plt.title('Linear Regression graph')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
plt.scatter(X,y,color='r',label='Scatter_values')
plt.plot(X,new_LR.predict(X_poly),color='b',label='line')
plt.title('Linear Regression graph')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```
