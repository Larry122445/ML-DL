```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
names  =['preg','plas','pres','skin','test','mass','pedi','age','class']
```

```python
dataset = pd.read_csv('C:\\Users\\ICT\\Desktop\\ML-DL\\Algorithms_from_Scratch\\DataPreparations\\pima-indians-diabetes.data.csv',names = names)
```

```python
dataset
```

histogram

```python
dataset.hist()
plt.subplots_adjust(hspace=0.5, wspace=0.5) # Adjust the horizontal and vertical spacing plt.show()
plt.show()
```

Density Plot

```python
dataset.plot(kind = 'density',subplots=True,layout=(3,3),sharex=False)
plt.subplots_adjust(hspace=0.5, wspace=0.5) # Adjust the horizontal and vertical spacing plt.show()
plt.show()
```

Box and whisker

```python
dataset.plot(kind = 'box',subplots=True,layout=(3,3),sharex=False,sharey=False)
plt.subplots_adjust(hspace=0.5, wspace=0.5) # Adjust the horizontal and vertical spacing plt.show()
plt.show()
```

correlations

```python
import numpy as np
```

```python
correlations = dataset.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

Scatter plot

```python
from pandas.plotting import scatter_matrix
scatter_matrix(dataset)
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```
