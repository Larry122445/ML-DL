# spacing around subplots

```python
fig3,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500),bins=50,color='black',alpha=0.5)
fig3.subplots_adjust(wspace=0,hspace=0)# use subplots adijust        
```

# Colors, Markers, and Line Styles

```python
fig = plt.figure()
ax_new  = fig.add_subplot()

ax_new.plot(np.random.standard_normal(30).cumsum(), color="red",         
linestyle="--", marker="*")
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data,color='red',linestyle='--',label='Default')
ax.plot(data,color='black',linestyle='--',drawstyle='steps-post',label='steps-post')
ax.legend()
```

# Setting the title, axis labels, ticks, and tick labels

```python
fig ,ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
```

```python
#xticks and xticklabels
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize=8)
```

```python
#add legends
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum(), color="black", label="one")
ax.plot(np.random.randn(1000).cumsum(), color="black", linestyle="dashed",label='two')
ax.legend()
```

# Annotations And Drawing On a Subplots

```python
#adding shape
fig, ax = plt.subplots()
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
color="green", alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
